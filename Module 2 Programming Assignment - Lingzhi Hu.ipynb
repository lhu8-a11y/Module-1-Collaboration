### 4.1 Conditional tests with if, elif, else

```python
secret = 8
guess = 4

if guess < secret:
    print("too low")
elif guess > secret:
    print("too high")
else:
    print("just right")
```

### 4.2 if/else statements

```python
small = True
green = True

if small and green:
    print("pea")
elif small and not green:
    print("cherry")
elif not small and green:
    print("watermelon")
elif not small and not green:
    print("pumpkin")
```

### 6.1 Use a for loop to print the values of the list

```python
for value in [3, 2, 1, 0]
    print(value)
```

### 6.2 While loop with comparison

```python
guess_me = 7
number = 1

while True:
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:
        print("oops")
        break
    number += 1
```

### 6.3 For loop with range()

```python
guess_me = 5

for number in range(10):
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:
        print("oops")
        break
```
