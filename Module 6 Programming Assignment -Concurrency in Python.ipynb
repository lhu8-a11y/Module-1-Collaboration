### 13.1 Write the current date as a string to the text file today.txt.

```python
import datetime

# get today's date
current_date = datetime.date.today()

# convert to iso format string
date_str = current_date.iosformat()

# write string to today.txt
with open('today.txt', 'w') as f:
    f.write(date_str)

print(f'Written to file: {date_str}')
```

### 13.2 Read the text file today.txt into the string today_string.

```python
with open('today.txt', 'r') as f:
    today_string = f.read().strip()

print(f'today_string content: {today_string}')
```

### 13.3 Parse the date from today_string.

```python
parsed_date = datetime.date.fromisoformat(today_string)

print(f'Parsed date object: {parsed_date}')
print(f'Type: {type(parsed_date)}')
print(f'Year: {parsed_date.year}, Month: {parsed_date.month}, Day: {parsed_date.day}')
```

### 15.1 Use multiprocessing to create three separate processes. Make each one wait a random number of seconds between zero and one, print the current time, and then exit.

```python
import multiprocessing
import random
import time
from datetime import datetime

def worker(process_name):
    # generate a random sleep time duration between 0 and 1 second
    sleep_time = random.uniform(0,1)
    time.sleep(sleep_time)

    # get current time format as YYYY-MM-DD HH:MM:SS, no microseconds
    now = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    
    # print process name, current time, and actual sleep duration
    print(f'{process_name} woke up, current time:{now} (slept {sleep_time:.3f} s)')

if __name__ == '__main__':
    # create and start 3 separate processes
    processes = []
    for i in range(3):
        p = multiprocessing.Process(
            target=worker,
            args=(f'Process-{i+1}',) # give each process a readable name
        )
        processes.append(p)
        p.start()

        for p in processes:
            p.join()
```
