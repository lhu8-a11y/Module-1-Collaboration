### Sort an Array of 0s, 1s, and 2s

```python
class Solution:
    def sort012(self, arr):
        # Initialize three pointers
        low, mid, high = 0, 0, len(arr) - 1

     # Traverse the array while mid pointer does not cross high pointer
        while mid <= high:
            if arr[mid] == 0:
                # Swap arr[low] and arr[mid], then move both pointers forward
                arr[low], arr[mid] = arr[mid], arr[low]
                low += 1
                mid += 1
            elif arr[mid] == 1:
                # 1 is already in correct region, just move mid forward
                mid += 1
            else:
                # Swap arr[mid] and arr[high], move high backward
                arr[mid], arr[high] = arr[high], arr[mid]
                high -= 1
                
if __name__ == "__main__":
    arr = [0, 1, 2, 0, 1, 2]
    n = len(arr)
    
    # sort the array in-place
    sol = Solution()
    sol.sort012(arr)

    for i in range(n):
        print(arr[i], end=' ')
```

### Binary Search

```python
class Solution:
    def binarysearch(self, arr, k):
        # Set left and right pointers
        low, high = 0, len(arr) - 1
        # Store the result, default -1
        result = -1
        
        # Binary search loop
        while low <= high:
            mid = (low + high) // 2
            
            # If middle element equals k
            if arr[mid] == k:
                result = mid
                high = mid - 1
            # If middle element is smaller, search right half
            elif arr[mid] < k:
                low = mid + 1
            # If middle element is larger, search left half
            else:
                high = mid - 1
        
        # Return the found index or -1
        return result
```
